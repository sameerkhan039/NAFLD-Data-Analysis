```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r'C:\Users\Hp\OneDrive\Apps\Desktop\project\nafld1.csv')
df2 = pd.read_csv(r'C:\Users\Hp\OneDrive\Apps\Desktop\project\nafld2.csv')
```

```python
print("Columns in df1:", df1.columns)
print("Columns in df2:", df2.columns)
```

```python
dffinal = pd.merge(df1 , df2 , on='id')
```

```python
col = ['id' , 'age', 'male','weight','height','bmi','case.id','futime','status','days','value']
for x in col:
    dffinal[x] = dffinal[x].fillna(dffinal[x].mean())
dffinal.isnull().sum()
```

```python
dffinal.describe()
```

```python
dffinal.to_csv(r'C:\Users\Hp\OneDrive\Apps\Desktop\project\cleaned_data.csv')
```

```python
plt.figure(figsize=(8, 5))
plt.hist(dffinal['value'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of All Test Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
insight1 = dffinal.groupby('test')[['value' , 'age' , 'bmi']].mean()
insight1
```
